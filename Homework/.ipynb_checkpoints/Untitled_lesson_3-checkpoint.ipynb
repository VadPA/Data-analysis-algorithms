```python
def loss_function(x):
    return 0 if x > 0 else 1
```

```python
dots = np.linspace(-3, 3, 1000)
q_zero_one_loss = [loss_function(x) for x in dots]

plt.xlabel('M')
plt.xlim(-3, 3)
plt.plot(dots, q_zero_one_loss);
```

```python
def exp_loss_func(x):
    return np.exp(-x)
```

```python
def logistic_loss(x):
    return np.log2(1 + np.exp(-x))
```

```python
q_exp_loss = [exp_loss_func(x) for x in dots]
q_logistic_loss = [logistic_loss(x) for x in dots]

plt.xlabel('M')
plt.xlim(-3, 3)
plt.ylim(0, 3)
plt.plot(dots, q_zero_one_loss)
plt.plot(dots, q_exp_loss)
plt.plot(dots, q_logistic_loss)
plt.legend(['zero-one loss', 'exponential loss', 'logistic loss']);
```

```python

#     sklearn.datasets
X, y = datasets.make_classification(n_samples=100, n_features=2, n_informative=2,
                                    n_redundant=0, n_classes=2, random_state=15)
# X, y = datasets.make_blobs(centers=2, cluster_std=2.5, random_state=12)
X, y
```

```python
#  
np.random.seed(12)
shuffle_index = np.random.permutation(X.shape[0])
X_shuffled, y_shuffled = X[shuffle_index], y[shuffle_index]

#      
train_proportion = 0.7
train_test_cut = int(len(X) * train_proportion)

X_train, X_test, y_train, y_test = \
    X_shuffled[:train_test_cut], \
    X_shuffled[train_test_cut:], \
    y_shuffled[:train_test_cut], \
    y_shuffled[train_test_cut:]
    
print("    ", X_train.shape)
print("    ", X_test.shape)
print("     ", y_train.shape)
print("     ", y_test.shape)
```

```python
def log_loss(w, X, y):
    m = X.shape[0]
    #   ,  
    A = sigmoid(np.dot(X, w))
        
    # labels 0, 1
    loss = -1.0 / m * np.sum(y * np.log(A) + (1 - y) * np.log(1 - A))
    
    # labels -1, 1
#     temp_y = np.where(y == 1, 1, -1)
#     loss = 1.0 / m * np.sum(np.log(1 + np.exp(-temp_y * np.dot(X, w))))

    grad = 1.0 / m * X.T @ (A - y)

    return loss, grad
```

```python
def optimize(w, X, y, n_iterations, eta):
    #          
    losses = []
    
    for i in range(n_iterations):        
        loss, grad = log_loss(w, X, y)
        w = w - eta * grad

        losses.append(loss)
        
    return w, losses
```

```python
def predict(w, X):
    
    m = X.shape[0]
    
    y_predicted = np.zeros(m)

    A = np.squeeze(sigmoid(np.dot(X, w)))

    #           0.5
    for i in range(A.shape[0]):
        if (A[i] > 0.5): 
            y_predicted[i] = 1
        elif (A[i] <= 0.5):
            y_predicted[i] = 0

    return y_predicted
```

```python
#    
w0 = np.zeros(X_train.shape[1])

n_iterations = 1000
eta = 0.05

w, losses = optimize(w0, X_train, y_train, n_iterations, eta)

y_predicted_test = predict(w, X_test)
y_predicted_train = predict(w, X_train)

#        
train_accuracy = np.mean(y_predicted_train == y_train) * 100.0
test_accuracy = np.mean(y_predicted_test == y_test) * 100.0

print(f"   w: {w}")
print(f"   : {train_accuracy:.3f}")
print(f"   : {test_accuracy:.3f}")
```

```python
plt.title('Log loss')
plt.xlabel('iterations')
plt.ylabel('loss')
plt.plot(range(len(losses)), losses);
```

### *1.   calc_logloss ,        np.log.

```python
def calc_logloss(y, y_pred):
    list_err = []
    for el in range(y.shape[0]):
        if (y_pred[el] == 0) | (y_pred[el] == 1):
            list_err.append(int(1) if y_pred[el] == 1 else int(0))
        else:
            err = - (y[el] * np.log(y_pred[el]) + (1.0 - y[el]) * np.log(1.0 - y_pred[el]))
            list_err.append(err)
    return s.mean(list_err)
```

```python
#  
y1 = np.array([1, 0])
y_pred1 = np.array([0.99, 0.99])
calc_logloss(y1, y_pred1)
```

```python
#   
y1 = np.array([1, 0])
y_pred1 = np.array([1, 0])
calc_logloss(y1, y_pred1)
```

### 2.    eval_model     ,  log loss  .

```python
def eval_model(X, y, iterations, eta=1e-4):
    np.random.seed(42)
    W = np.random.randn(X.shape[1])
    n = X.shape[0]
    
    for i in range(iterations):
        z = np.dot(X, W)
        y_pred = sigmoid(z)
        err = calc_logloss(y, y_pred)
        
        dQ = 1/n * X.T @ (y_pred - y)
        W -= eta * dQ
        if (i % (iterations / 10) == 0) | (i == iterations-1) | (i == iterations-2) | (i == iterations-3):
            print(i, W, err)
    return W
```

```python
W = eval_model(X_st, y, iterations=38000, eta=1e-1)
```

```python
W2 = eval_model(X_st, y, iterations=20000, eta=1e-1)
```
